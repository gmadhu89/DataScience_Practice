```python
import numpy as np
from scipy import stats
import requests
import csv

%matplotlib inline
import seaborn as sns
```

```python
r = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
r.encoding = 'utf-8'

iris_data = [str.split(row,",") for row in r.iter_lines(decode_unicode=True)]

print("First five entries in Iris data is \n")
for i in iris_data[0:5]:
    print(i)
    
print("Last five entries in iris data is \n")
for i in iris_data[-5:]:
    print(i)
    
#Cleansing the data 
iris_cleansed = iris_data[:-1]

#Names in a separate list
iris_names = [names[-1] for names in iris_cleansed]
#Values in a separate list
iris_values = [values[0:-1] for values in iris_cleansed]
iris_values[0:5]
#Convert all values to floats
iris_values_float = [[float(values) for values in num] for num in iris_values]
iris_values_float[0:5]
# Create a matrix now

iris_data_final = np.array(iris_values_float)
print(iris_data_final,iris_data_final.shape)
```

```python

print("Mean of each column is : ", iris_data_final.mean(axis=0))
print("Median of each column is : ", np.median(iris_data_final,axis=0))
print("Mode of each column is : ", stats.mode(iris_data_final,axis=0))

print("Variance of each column is :",iris_data_final.var(axis=0))
print("Std Deviation of each column is :",iris_data_final.std(axis=0))

#Covariance of iris_data_final
iris_data_final_column_cov = np.cov(iris_data_final.T)
print("Covariance of Iris data final set is ",iris_data_final_column_cov)
print("Original Iris data column matrix shape is ",iris_data_final.shape)
print("Covariance matrix is ",iris_data_final_column_cov,iris_data_final_column_cov.shape)
```

```python
print("Columnar Covariance matrix is \n",iris_data_final_column_cov,iris_data_final_column_cov.shape)
```

```python
print("Row based covariance matrix of iris_data n")
iris_data_final_row_cov = np.cov(iris_data_final)
print(iris_data_final_row_cov,iris_data_final_row_cov.shape)
```

```python
iris_data_final_column_corr = np.corrcoef(iris_data_final.T)
print("COlumnar Correlation matrix is \n",iris_data_final_column_corr,iris_data_final_column_corr.shape)
```

```python
sns.heatmap(iris_data_final_column_corr)
```
